```python
import numpy as np
from matplotlib import pyplot as plt
```

### Confidence level

```python
c = 0.95
```

### Probabilities of dice results

```python
p = np.concatenate((np.arange(1, 7), np.arange(5, 0, -1))) / 36

plt.plot(np.arange(2, 13), p*100, 'o--')
plt.title('Probability of rolling a sum of 2 dice')
plt.xlabel('Dice sum')
plt.ylabel('Probability / %')
plt.grid()
```

### Number of required trials to get desired result

```python
Navg = 1/p
Nmax = np.log(1 - c) / np.log(1 - p)
```

```python
x = np.arange(2, 13)

plt.plot(x, Navg, 'o--', label= 'Average number of trials')
plt.plot(x, Nmax, 'o--', label=f'Maximum number of trials, {c:.0%} confidence level')
plt.title('Number of required trials to roll a sum of 2 dice')
plt.xlabel('Dice sum')
plt.ylabel('Number of trials')
plt.legend()
plt.grid()

Navg
```

### Game points in relation to number of crosses in one line

```python
n = np.arange(1,12)

Points = n*(n+1)/2

plt.plot(n, Points, 'o--')
plt.title('Game points over number of crosses in one line')
plt.xlabel('Number of crosses')
plt.ylabel('Points')
plt.grid()

Points
```

### Relationship between game points and required trials

Since more trials are required to get extremal values, the optimal strategy is to only fill the central boxes.  
E.g. to get 3 crosses (= 6 points) the player tries to fill 6, 7 & 8, for which on average 7 + 6 + 7 = 20 dice rolls would be needed.

```python
Nmin_avg = np.cumsum(np.sort(Navg))

plt.plot(Nmin_avg, Points, 'o--', label= 'Average number of trials')
plt.title('Average required trials (optimal) for desired game points')
plt.xlabel('Number of trials')
plt.ylabel('Points')
plt.grid()

for x, y, text in zip(Nmin_avg[:-1], Points[:-1], np.arange(1, 11)):
    plt.text(x, y, text)
plt.text(Nmin_avg[-1] - 20, Points[-1], '11 crosses')

Nmin_avg
```
