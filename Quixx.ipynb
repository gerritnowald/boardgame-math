```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
c = 0.95    # confidence level
```

```python
# probabilities of dice results
p = np.concatenate((np.arange(1, 7), np.arange(5, 0, -1))) / 36

plt.plot(np.arange(2, 13), p*100, 'o--')
plt.title('Probability of rolling a sum of 2 dice')
plt.xlabel('Dice sum')
plt.ylabel('Probability / %')
plt.grid()
```

```python
# number of necessary trials
Navg = 1/p
Nmax = np.log(1 - c) / np.log(1 - p)
```

```python
x = np.arange(2, 13)

plt.plot(x, Navg, 'o--', label= 'Average number of trials')
plt.plot(x, Nmax, 'o--', label=f'Maximum number of trials, {c:.0%} confidence level')
plt.title('Number of required trials to roll a sum of 2 dice')
plt.xlabel('Dice sum')
plt.ylabel('Number of trials')
plt.legend()
plt.grid()
```

```python
n = np.arange(1,12)

Points = n*(n+1)/2

plt.plot(n, Points, 'o--')
plt.title('Game points over number of crosses in one line')
plt.xlabel('Number of crosses')
plt.ylabel('Points')
plt.grid()

Points
```

```python
plt.plot(np.cumsum(np.sort(Navg)), Points, 'o--', label= 'Average number of trials')
plt.plot(np.cumsum(np.sort(Nmax)), Points, 'o--', label=f'Maximum number of trials, {c:.0%} confidence level')
plt.title('Minimal required trials for desired game points')
plt.xlabel('Number of trials')
plt.ylabel('Points')
plt.legend()
plt.grid()
```
